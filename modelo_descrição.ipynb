Vamo criar um modelo baseado apenas nas descrições dos jogos vindas da base "steam_description_data"

```python
import pandas as pd

descriptions = pd.read_csv("archive\steam_description_data.csv")
descriptions
```

```python
game_data = pd.read_csv("archive\steam.csv")
game_data
```

```python
game_data = game_data.rename(columns={"appid":"app_id"})
game_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
tf_idf_vectorizer = TfidfVectorizer()
def similarity(text_lst):
    tf_idf_transform = tf_idf_vectorizer.fit_transform(text_lst)
    return cosine_similarity(tf_idf_transform)

desc_list = list(descriptions["detailed_description"])
sim_list = similarity(desc_list)
```

```python
sim_df = pd.DataFrame(sim_list)
sim_df
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=6)
clusters = k_means.fit_predict(sim_list)
```

```python
from matplotlib import pyplot as plt
import numpy as np
lookup_recomendation = {"app_id":  descriptions["steam_appid"], "cluster": clusters}
lookup_recomendation = pd.DataFrame(lookup_recomendation)

unique_clusters = np.unique(clusters)
for i in unique_clusters:
    plt.scatter(lookup_recomendation.loc[lookup_recomendation["cluster"]==i]["cluster"].sample(100),lookup_recomendation.loc[lookup_recomendation["cluster"]==i]["app_id"].sample(100) )
```

```python
lookup_recomendation = pd.merge(game_data.iloc[:, 0:2], lookup_recomendation, on = "app_id")
lookup_recomendation
```

Agora vamos testar o modelo

```python
from random import sample

def get_recomendation(game):
    cluster_to_recommend = int(lookup_recomendation.loc[lookup_recomendation["name"]==game]["cluster"])
    return sample(list(lookup_recomendation.loc[lookup_recomendation["cluster"]==cluster_to_recommend]["name"]),5)

get_recomendation("Team Fortress Classic")
```
