Vamo criar um modelo baseado apenas nas descrições dos jogos vindas da base "steam_description_data"

```python
import pandas as pd

descriptions = pd.read_csv("archive/steam_description_data.csv")
descriptions
```

```python
game_data = pd.read_csv("archive/steam.csv")
game_data
```

```python
game_data = game_data.rename(columns={"appid":"app_id"})
game_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
tf_idf_vectorizer = TfidfVectorizer()
def similarity(text_lst):
    tf_idf_transform = tf_idf_vectorizer.fit_transform(text_lst)
    return cosine_similarity(tf_idf_transform)

desc_list = list(descriptions["detailed_description"])
sim_list = similarity(desc_list)
```

```python
sim_df = pd.DataFrame(sim_list)
sim_df
```

Agora vamos testar o modelo

A matriz "sim_df" contem as similaridades entre o jogo i (linha) e o jogo j (coluna) </br>
então, vamos realizar uma recomendação baseada em quais jogos j estão mais proximos de 1 para um jogo i

```python

def get_recomendation_sorting(game):
    game_index = game_data.index[game_data["name"]==game].to_list()[0]
    game_i = sim_df.iloc[game_index].sort_values(ascending=False)
    recomendation_index = game_i.iloc[1:6].index
    return game_data.iloc[recomendation_index]
    

pd.DataFrame(get_recomendation_sorting("Counter-Strike: Global Offensive"))
```
